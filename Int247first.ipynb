```
import numpy as np
import matplotlib.pyplot as plt
```

```
a=np.arange(-5,5,0.2)
```

```
len(a)
```

```
print(a)
```

```
y1=a
```

```
plt.plot(a,y1)
plt.title("Graph 1")
plt.xlabel("A")
plt.ylabel("Y1")
plt.show()
```

```
y2=2*a+5
```

```
plt.plot(a,y2)
plt.title("graph2")
plt.xlabel("a")
plt.ylabel("y2")
plt.show()
```

```
y3=a**2
```

```
plt.plot(a,y3)
plt.title("graph3")
plt.show()
```

```
y4=4*a**3+a**2+4*a+7
```

```
plt.plot(a,y4)
plt.title("graph4")
plt.show()
#x and y axes are same for all
```

```
y5=1/(1+np.exp(-a))
```

```
plt.plot(a,y5)
plt.title("Graph5")
plt.show()
```

```
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(a,y1) 
plt.title("Graph1") 
plt.subplot(2,2,2)
plt.plot(a,y3)
plt.title("Graph3")
plt.subplot(2,2,3)
plt.plot(a,y4)
plt.title("Graph4")
plt.subplot(2,2,4)
plt.plot(a,y5)
plt.title("Graph5")
plt.show()
```
